```python
# Function to count frequency
def tuple_freq(t):
    freq = {}
    for i in t:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    return freq

# Main
print(tuple_freq((1,2,2,3,1,4)))
```
