```python
# Function to find duplicate elements
def find_duplicates(lst):
    duplicates = []

    # Loop through list
    for i in lst:
        # Check if element appears more than once
        if lst.count(i) > 1 and i not in duplicates:
            duplicates.append(i)

    return duplicates

# Main program
lst = [1, 2, 3, 2, 4, 5, 1]
result = find_duplicates(lst)

print("Duplicate elements:", result)
```
