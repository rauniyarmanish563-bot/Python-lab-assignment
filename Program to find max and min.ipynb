```python
# Function to find max and min
def find_max_min(t):
    max_val = t[0]
    min_val = t[0]

    for i in t:
        if i > max_val:
            max_val = i
        if i < min_val:
            min_val = i

    return max_val, min_val

# Main
t = (5,1,8,3)
mx, mn = find_max_min(t)
print("Max =", mx, "Min =", mn)
```
